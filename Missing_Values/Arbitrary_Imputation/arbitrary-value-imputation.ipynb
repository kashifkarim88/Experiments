# Arbitrary Value Imputation

- **We use *Arbitrary Value Imputation* when the data is not missing at random**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('titanic_toy.csv')
```

# End of Distribution Imputation

- **It is hard to find the best arbitrary value therefore we use End of Distribution Imputation to find the best arbitrary value.**

- **we use it when data is not missing at random.**
- **Here we use 3 segma rule if the distribution is normal i.e. (mean - 3 segma) or (mean + 3 segma) and if the distribution is skewed we use IQR Proximity to get the end value of distribution.**

```python
df.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns = ['Age', 'Fare']

for col in columns:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], kde=True)
    plt.title(f"{col} Distribution (Before Imputation)")
    plt.show()
```

```python
age = abs(df['Age'].skew())
fare = abs(df['Fare'].skew())
age, fare
```

```python
end_values = {}

for col in columns:
    skewness = df[col].dropna().skew()
    
    print(f"\nColumn: {col}")
    print(f"Skewness: {skewness:.3f}")

    if abs(skewness) < 0.5:
        mean = df[col].dropna().mean()
        std = df[col].dropna().std()
        end_value = mean + 3 * std
        method = "3-Sigma"
        
        print(f"Mean: {mean:.2f}")
        print(f"Std: {std:.2f}")
        print(f"End Value (Mean + 3): {end_value:.2f}")

    else:
        Q1 = df[col].dropna().quantile(0.25)
        Q3 = df[col].dropna().quantile(0.75)
        IQR = Q3 - Q1
        end_value = Q3 + 1.5 * IQR
        method = "IQR Proximity"

        print(f"Q1: {Q1:.2f}")
        print(f"Q3: {Q3:.2f}")
        print(f"IQR: {IQR:.2f}")
        print(f"End Value (Q3 + 1.5IQR): {end_value:.2f}")

    print(f"Method Used: {method}")
    end_values[col] = end_value
```

```python
end_values
```

```python
for col in columns:
    df[f"{col}_end_val"] = df[col].fillna(end_values[col])
```

```python
df.head()
```

```python
for col in columns:
    fig, axes = plt.subplots(1, 2, figsize=(12,4))
    
    if col == 'Age':
        sns.histplot(df[col], kde=True, ax=axes[0])
        axes[0].set_title(f"{col} Before Imputation")

        sns.histplot(df[col], kde=True, ax=axes[1])
        axes[1].axvline(end_values[col], color='red', linestyle='--')
        axes[1].set_title(f"{col} After Imputation")
        plt.show()
    else:
        sns.histplot(df[col],bins=25, kde=True, ax=axes[0])
        axes[0].set_title(f"{col} Before Imputation")
        
        sns.histplot(df[col],bins=25, kde=True, ax=axes[1])
        axes[1].axvline(end_values[col], color='red', linestyle='--')
        axes[1].set_title(f"{col} After Imputation")
        plt.show()
```

**End of Distribution Imputation is a missing-value handling technique designed for situations where data is not missing at random (NMAR) and the absence of values itself carries information.**

*Instead of replacing missing values with central tendencies such as the mean or median which can distort the original data distribution this method imputes missing entries using a value located at the extreme end of the features distribution.*

The choice of this end value depends on the shape of the data: for approximately normally distributed features, the 3-sigma rule is applied by selecting a value equal to the mean plus three standard deviations, ensuring the imputed values lie far from the center while remaining statistically valid. 

For skewed distributions, where the normality assumption does not hold, the IQR proximity method is used, typically defined as the third quartile plus 1.5 times the interquartile range, which places imputed values beyond the upper spread of the data. 

By pushing missing values to the distribution tail, End of Distribution Imputation preserves the original distribution shape, maintains variance, and allows models especially tree-based and linear models to learn the potential significance of missingness without introducing bias toward the mean.
